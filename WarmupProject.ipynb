```python
import gym

# Create the environment with modern render mode
env = gym.make('CartPole-v1', render_mode="human")

for i_episode in range(20):
    observation, info = env.reset()  # New API: reset returns (obs, info)
    
    for t in range(100):
        env.render()
        print(observation)
        
        action = env.action_space.sample()
        
        # New API: step returns 5 values
        observation, reward, terminated, truncated, info = env.step(action)
        
        if terminated or truncated:
            print(f"Episode finished after {t+1} timesteps")
            break

# Print env specs
print("\nAction space:", env.action_space)
print("Observation space:", env.observation_space)
print("Observation high:", env.observation_space.high)
print("Observation low:", env.observation_space.low)

env.close()
```
