```python
import random
import gym
import math
import numpy as np
from collections import deque
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

# ---------------------- Training Parameters ----------------------
n_episodes = 1000
n_win_ticks = 195
max_env_steps = None
gamma = 1.0
epsilon = 1.0
epsilon_min = 0.01
epsilon_decay = 0.995
alpha = 0.01
alpha_decay = 0.01
batch_size = 64
monitor = False
quiet = False

# ---------------------- Environment Setup ----------------------
memory = deque(maxlen=100000)
env = gym.make('CartPole-v1', render_mode="human")  # ✅ Updated env name and added render_mode

if max_env_steps is not None:
    env._max_episode_steps = max_env_steps  # ✅ Warning: underscore method is used here for older-style limit

# ---------------------- Model Definition ----------------------
model = Sequential()
model.add(Dense(24, input_dim=4, activation='relu'))
model.add(Dense(48, activation='relu'))
model.add(Dense(2, activation='linear'))  # ✅ Changed activation to 'linear' for Q-value output
model.compile(loss='mse', optimizer=Adam(learning_rate=alpha, decay=alpha_decay))  # ✅ 'lr' changed to 'learning_rate'

# ---------------------- Memory Functions ----------------------
def remember(state, action, reward, next_state, done):
    memory.append((state, action, reward, next_state, done))

def choose_action(state, epsilon):
    if np.random.random() <= epsilon:
        return env.action_space.sample()
    q_values = model.predict(state, verbose=0)  # ✅ Suppressed verbose output
    return np.argmax(q_values)

def get_epsilon(t):
    return max(epsilon_min, min(epsilon, 1.0 - math.log10((t + 1) * epsilon_decay)))

def preprocess_state(state):
    return np.reshape(state, [1, 4])

def replay(batch_size, epsilon):
    x_batch, y_batch = [], []
    minibatch = random.sample(memory, min(len(memory), batch_size))
    for state, action, reward, next_state, done in minibatch:
        y_target = model.predict(state, verbose=0)
        if done:
            y_target[0][action] = reward
        else:
            future_q = model.predict(next_state, verbose=0)[0]
            y_target[0][action] = reward + gamma * np.max(future_q)

        x_batch.append(state[0])
        y_batch.append(y_target[0])

    model.fit(np.array(x_batch), np.array(y_batch), batch_size=len(x_batch), verbose=0)

# ---------------------- Main Loop ----------------------
def run():
    scores = deque(maxlen=100)

    for e in range(n_episodes):
        state, _ = env.reset()  # ✅ Updated for Gym >= 0.26
        state = preprocess_state(state)
        done = False
        i = 0

        while not done:
            action = choose_action(state, get_epsilon(e))
            next_state, reward, terminated, truncated, _ = env.step(action)  # ✅ Updated to handle new step API
            done = terminated or truncated  # ✅ Combine terminated and truncated
            env.render()

            next_state = preprocess_state(next_state)
            remember(state, action, reward, next_state, done)
            state = next_state
            i += 1

        scores.append(i)
        mean_score = np.mean(scores)

        if mean_score >= n_win_ticks and e >= 100:
            if not quiet:
                print(f'Ran {e} episodes. Solved after {e - 100} trials')
            return e - 100

        if e % 20 == 0 and not quiet:
            print(f'[Episode {e}] - Mean survival time over last 100 episodes: {mean_score:.2f} ticks')

        replay(batch_size, get_epsilon(e))

    if not quiet:
        print(f'Did not solve after {n_episodes} episodes')
    return n_episodes

run()

```
